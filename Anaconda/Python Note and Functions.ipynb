```python
def anagram2(s1, s2):
    """To check if the words are anagram"""
    
    """Remove spaces"""
    s1 = s1.replace(" ", "").lower()
    s2 = s2.replace(" ", "").lower()
    
    """create a dictionary to check the count of the digits"""
    count = {}
    
    """check if letter is in count, if yes, add 1 to value, if no add key and value 1"""
    for letter in s1:
        if letter in count:
            count[letter] += 1
        else:
            count[letter] = 1
    
    """check if letter is in count, if yes, subtract 1 to value, if no add key and value 1"""
    for letter in s2:
        if letter in count:
            count[letter] -= 1
        else:
            count[letter] = 1
    """check key and value in count. if it's not 0, return false, else true"""       
    for k in count:
        print(count[k])
        if count[k] == 0:
            return True
        else:
            return False
```

```python
def large_cont_sum(arr): 
    
    # Check to see if array is length 0
    if len(arr)==0: 
        return 0
    
    # Start the max and current sum at the first element
    max_sum=current_sum=arr[0] 
    
    # For every element in array
    for num in arr[1:]: 
        
        # Set current sum as the higher of the two
        current_sum=max(current_sum+num, num)
        
        # Set max as the higher between the currentSum and the current max
        max_sum=max(current_sum, max_sum) 
        
    return max_sum 
```

```python
def rev_word3(s):
    """
    Manually doing the splits on the spaces.
    """
    
    words = []
    length = len(s)
    spaces = [' ']
    
    # Index Tracker
    i = 0
    
    # While index is less than length of string
    while i < length:
        
        # If element isn't a space
        if s[i] not in spaces:
            
            # The word starts at this index
            word_start = i
            
            while i < length and s[i] not in spaces:
                
                # Get index where word ends
                i += 1
            # Append that word to the list
            words.append(s[word_start:i])
        # Add to index
        i += 1
        
    # Join the reversed words
    return " ".join(reversed(words))
```

```python
def compress(s):
    """
    This solution compresses without checking. Known as the RunLength Compression algorithm.
    """
    
    # Begin Run as empty string
    r = ""
    l = len(s)
    
    # Check for length 0
    if l == 0:
        return ""
    
    # Check for length 1
    if l == 1:
        return s + "1"
    
    #Intialize Values
    last = s[0]
    cnt = 1
    i = 1
    
    while i < l:
        
        # Check to see if it is the same letter
        if s[i] == s[i - 1]: 
            # Add a count if same as previous
            cnt += 1
        else:
            # Otherwise store the previous data
            r = r + s[i - 1] + str(cnt)
            cnt = 1
            
        # Add to index count to terminate while loop
        i += 1
    
    # Put everything back into run
    r = r + s[i - 1] + str(cnt)
    
    return r
```

```python
def balance_check(k):
    
    if len(k) % 2 == 0:
        return True
    
    opening_matches = set('({[')
    
    matches = set([('(',')'), ("[","]"), ("{","}")])
    
    stack = []
    
    for paren in k:
        if paren in opening_matches:
            stack.append(paren)
        else:
            if len(stack) == 0:
                return False
            
            last_open = stack.pop()
            
            if(last_open, paren) not in matches:
                return False
            
    return len(stack) == 0
```

```python
class Deque(object):
    
    """Deque does not follow neither Stack or Queue ordering. It can add items from either ends (Start or END)"""
    
    def __init__(self):
        """Create empty list"""
        self.items = []
        
    def isEmpty(self):
        """Return True if list is empty"""
        return self.items == []
    
    def addFront(self, item):
        """Add items in the list from the front"""
        self.items.append(item)
    
    def addRear(self, item):
        """Add items in the list from the back"""
        self.items.insert(0,item)
        
    def removeFront(self):
        """Remove items in the list from the front"""
        return self.items.pop()
    
    def removeRear(self):
        """Remove items in the list from the back"""
        return self.items.pop(0)
    
    def size(self):
        """Return the lenght of the list"""
        return len(self.items)
```

```python
class Queue3Stacks(object):
    
    def __init__(self):
        
        # Two Stacks
        self.instack = []
        self.outstack = []
     
    def enqueue(self,element):
        
        # Add an enqueue with the "IN" stack
        self.instack.append(element)
    
    def dequeue(self):
        if not self.outstack:
            while self.instack:
                # Add the elements to the outstack to reverse the order when called
                self.outstack.append(self.instack.pop())
        return self.outstack.pop()  
```

```python
def cycle_check(node):

    # Begin both markers at the first node
    marker1 = node
    marker2 = node

    # Go until end of list
    while marker2 != None and marker2.nextnode != None:
        
        # Note
        marker1 = marker1.nextnode
        marker2 = marker2.nextnode.nextnode

        # Check if the markers have matched
        if marker2 == marker1:
            return True

    # Case where marker ahead reaches the end of the list
    return False
```

```python
def word_split(phrase,list_of_words, output = None):
    '''
    Note: This is a very "python-y" solution.
    ''' 
    
    # Checks to see if any output has been initiated.
    # If you default output=[], it would be overwritten for every recursion!
    if output is None:
        output = []
    
    # For every word in list
    for word in list_of_words:
        
        # If the current phrase begins with the word, we have a split point!
        if phrase.startswith(word):
            
            # Add the word to the output
            output.append(word)
            
            # Recursively call the split function on the remaining portion of the phrase--- phrase[len(word):]
            # Remember to pass along the output and list of words
            return word_split(phrase[len(word):],list_of_words,output)
    
    # Finally return output if no phrase.startswith(word) returns True
    return output
```

```python
def permute(s):
    out = []
    
    # Base Case
    if len(s) == 1:
        out = [s]
        
    else:
        # For every letter in string
        for i, let in enumerate(s):

            
            # For every permutation resulting from Step 2 and 3 described above
            for perm in permute(s[:i] + s[i+1:]):
                print(permute(s[:i]) + s[i+1:]) 
                
                # Add it to output
                out += [let + perm]

    return out
```

```python
def recur(b):
    
    if b == 0:
        return 1
    
    return b * recur(b-1)


recur(10)
```

```python
def recur_iter(b):
    
    factor = 1
    
    for i in range(1, b+1):
        factor *= i
    return factor

recur_iter(5)
```

```python
# Program to display the Fibonacci sequence up to n-th term

nterms = int(input("How many terms? "))

# first two terms
n1, n2 = 0, 1
count = 0

# check if the number of terms is valid
if nterms <= 0:
    print("Please enter a positive integer")
elif nterms == 1:
    print("Fibonacci sequence upto",nterms,":")
    print(n1)
else:
    print("Fibonacci sequence:")
while count < nterms:
    print(n1)
    nth = n1 + n2
    # update values
    n1 = n2
    n2 = nth
    count += 1
```

```python
def rotate_left3(nums):
  t = nums[0]
  b = nums[1:]
  
  return int(b)+int(t)

rotate_left3([1,2,3])
```

```python
def count_hi(str):
  list = []
  for x  in range(len(str)):
      if str[x] == "h" and str[x+1] == "i":
          t = str[x] +str[x+1]
          list.append(t)
          
  return len(list)
count_hi("asdasdashi, hi")
```

```python
def count_hi2(s):
    count = 0
    for i in range(len(s)-1):
        print(i)
        count += s[i:i+2] == 'hi'
    return count

count_hi2("asdashihihi")
```

```python
def prime_number_factor(nums):
    factor = list()
    divisor = 2
    
    while (divisor <= nums):
        if (nums%divisor == 0):
            factor.append(divisor)
            nums = nums/divisor
        else:
            divisor += 1
    return factor

prime_number_factor(56)
```

```python
import re

def is_palindrome(phrase):
    forward = ''.join(re.findall(r'[a-z]+', phrase.lower()))
    backward = forward[::-1]
    return forward == backward
is_palindrome("wee eew")
```
